## Task 1: Prediction using Supervised ML


### Question : Predict the percentage of an student based on the no. of study hours.


DataSet link: http://bit.ly/w-data


### Import the requried libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

### Load the requried dataset

```python
df = pd.read_csv('https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv')
df
```

### Data virtualization

```python
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Scores based on study hours')
plt.scatter(df['Hours'], df['Scores'])
plt.show()
```

### Extracting the Input and Target variables

```python
X = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values
```

### Splitting the training and testing datasets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

### Creating the supervised machine learning model
From the above data virtualization, we can clearly see that the, input(Hours) and target(Scores) are Linear in nature, So we use the LinearRegression model from sklearn library 

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

### Predicting the output from our trained model

What is the score of the student, having the study hour of 9.25 hours?

```python
model.predict([[9.25]])
```

### Accuracy of the model

```python
model.score(X_test, y_test)
```

### LinearRegression line for the model

```python
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Score based on study hours')
plt.scatter(df['Hours'], df['Scores'])
plt.plot(df['Hours'], model.predict(df[['Hours']]), color = 'red')
plt.show()
```

```python
y_test
```

```python
y_pred = model.predict(X_test)
y_pred
```

### Actual output vs Predicted ouput

```python
ans = pd.DataFrame({'Actual' : y_test, 'Predicted' : y_pred})
```
